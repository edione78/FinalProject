```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = "Resources/chopped_dataset2016.csv"
data2 = "Resources/population.csv"
```

```python
yearly = pd.read_csv(data)
yearly_df = pd.DataFrame(yearly)
yearly_df.head()
```

```python
grouped_yearly_df = yearly_df.groupby(['Country Name'])

grouped_yearly_df.count().head(10)
```

```python
ConstantAmount_df = pd.DataFrame(
    grouped_yearly_df["constant_amount"].sum())

ConstantAmount_df.head(10)
```

```python
population = pd.read_csv(data2)
df = pd.DataFrame(population)
df.head()
```

```python
df_filtered = df[['Country Name', '2016']]

df_filtered
```

```python
merged_df = pd.merge(df_filtered, ConstantAmount_df, how = 'left', 
                     on=['Country Name'])
merged_df
```

```python
df = merged_df.dropna(axis=0, subset=['constant_amount'])
df
```

```python
df.to_csv('2016_AID.csv')
```
