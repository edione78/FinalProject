```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = "Resources/population.csv"
```

```python
population = pd.read_csv(data)
df = pd.DataFrame(population)
df.head()
```

```python
df_filtered = df[['Country Name', 'Country Code', 'Indicator Name',
                   '2015', '2016', '2017', '2018', '2019']]
df_filtered
```

```python
df_population = df_filtered.rename(columns = {"Country Code":"country_code",
                                        "Country Name": 'country_name',
                                        "Indicator Name": 'indicator_name'})
df_population
```

```python
df_population.to_csv ('population_filtered.csv', index=False)
```

```python
filepath = "Resources/cleaned_dataset.csv"
dataset = pd.read_csv(filepath)
dataset
```

```python
merged_df = pd.merge(dataset, df_population, how = 'left', 
                     on=['country_code', 'country_name'])
merged_df
```

```python
merged_df.to_csv('merged_dataset.csv')
```
